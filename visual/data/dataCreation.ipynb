```python
import pandas as pd
import numpy as np
import json
import datetime
import random
```

```python
with open("superedge_locations_dates.json", 'r') as f:
    dates = json.load(f)
```

```python
dates['publicationDate']
```

```python
def time_converter(time_stamp):
    normalizer = 100000;
    return datetime.datetime.fromtimestamp(time_stamp/normalizer).strftime('%Y')
```

```python
new_dates = {k: time_converter(v) for k, v in dates['publicationDate'].items()}
```

```python
new_dates
```

```python
with open("superedge_locations.json", 'r') as f:
    connections = json.load(f)
```

```python
connections
```

```python
with open("newloc2coord.json", 'r') as f:
    standart_names = json.load(f)
```

```python
standart_names
```

```python
canton_names = list()
for row in pd.read_json("ch.json")['features']:
    canton_names.append(row['properties']['name'])
```

```python
standartized_connections = list()
for i  in connections:
    try:
        standartized_connections.append([standart_names[loc] for loc in connections[i]])
    except:
        pass
```

```python
dummy_list = list()
for index, l in enumerate(standartized_connections):
    connect_list = []
    for loc in l:
        if loc !=None and loc in canton_names and loc not in connect_list :
            connect_list.append(loc)
    dummy_list.append((connect_list,new_dates[list(connections.keys())[index]]))
```

```python
dummy_list
```

```python
cleaned_connections = dict()
for l in standartized_connections:
    for loc in l:
        if loc:
            cleaned_connections[loc] = dict()
```

```python
# IMPROVE > MAKE AUTOMATIC
dates = ['1970', '1971', '1972']
for d in dates:
    for k in cleaned_connections:
        cleaned_connections[k][d] = [];
```

```python
'''
# Get the Date from Real Data
for l in dummy_list:
    if len(l[0]) > 2:
        for loc in l[0]:
            cleaned_connections[loc][l[1]].append(l[0])
'''
```

```python
# Assign a Random Date
for l in dummy_list:
    if len(l[0]) > 2:
        for loc in l[0]:
            cleaned_connections[loc][random.choice(dates)].append(l[0])
```

```python
resulting_connections = dict()
for loc in cleaned_connections:
    if loc in canton_names:
        resulting_connections[loc] = cleaned_connections[loc]
```

```python
with open('canton_connections.js', 'w') as outfile:
    json.dump(resulting_connections, outfile, ensure_ascii=False)
```

### `Canton Locations`

```python
canton_coordinates = dict()
for canton in pd.read_json("ch.json")['features']:
    df = pd.DataFrame(canton['geometry']['coordinates'][0][0])
    canton_coordinates[canton['properties']['name']] = (df.mean(axis=0)[1].round(5), df.mean(axis=0)[0].round(5))
```

```python
with open('canton_coordinates.js', 'w') as outfile:
    json.dump(canton_coordinates, outfile, ensure_ascii=False)
```
